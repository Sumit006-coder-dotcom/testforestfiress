## Algerian Forest Fires Dataset

Data Set Information

The dataset includes 244 instances that regroup data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period covered is from June 2012 to September 2012.

The dataset includes 11 attributes and 1 output attribute (class).

The 244 instances have been classified into:

Fire (138 instances)

Not fire (106 instances)

Attribute Information:

1. Date: (DD/MM/YYYY) Day, month (June to September), year (2012)


2. Temp: Temperature at noon (max temperature) in Celsius degrees (22 to 42)


3. RH: Relative Humidity in % (21 to 90)


4. WS: Wind speed in km/h (6 to 29)


5. Rain: Total rainfall in mm (0 to 16.8)


6. Fine Fuel Moisture Code (FFMC): Index from the FWI system (28.6 to 92.5)


7. Duff Moisture Code (DMC): Index from the FWI system (1.1 to 65.9)


8. Drought Code (DC): Index from the FWI system (7 to 220.4)


9. Initial Spread Index (ISI): Index from the FWI system (0 to 18.5)


10. Buildup Index (BUI): Index from the FWI system (1.1 to 68)


11. Fire Weather Index (FWI): Index (0 to 31.1)


12. Classes: Two classes - Fire and Not Fire

## Data Cleaning

```python
## Missing Values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index. we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum() 
```

```python
## Removibg the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
##Remove the 122nd row
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns names
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

#### Changes the required columns as integer data type

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

#### Changing the other columns  to float datatypes

```python
objects = [features  for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```
